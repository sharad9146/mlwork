```python
df_score=df.drop(columns=["ever_married","work_type","Residence_type","smoking_status"])
```

```python
mean_value=df_score['bmi'].mean()
```

```python
# splitting the data in x and y that is input and output
x=df_score.iloc[:,2:11]
y=df_score['stroke'].values
```

```python
# splitting the data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# an initial SVM model with linear kernel
clf=SVC(kernel="linear",random_state=0)
```

```python
# fill the model
clf.fit(x_train,y_train)
```
